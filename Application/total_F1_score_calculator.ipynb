```python
import json
import re

with open("F1_score_calculator.ipynb") as file:
    a = json.load(file)

print(a)

total_true_positives = 0
total_false_positives = 0
total_true_negatives = 0
total_false_negatives = 0

for i in a['cells']:
    n = i['outputs'][0]['text']
    for j in n:
        if "True positives: " in j:
            total_true_positives += int(re.search(r"\d+", j)[0])
        if "False positives: " in j:
            total_false_positives += int(re.search(r"\d+", j)[0])
        if "True negatives: " in j:
            total_true_negatives += int(re.search(r"\d+", j)[0])
        if "False negatives: " in j:
            total_false_negatives += int(re.search(r"\d+", j)[0])

precision = total_true_positives/(total_true_positives + total_false_positives)
recall = total_true_positives/(total_true_positives + total_false_negatives)

f1_score = (2*precision*recall)/(precision+recall)
print(f1_score)
```
