```python
import json
import re
import ast

pattern = r"\[.*\]"
with open("F1_score_calculator.ipynb") as file:
    jsonContent = json.load(file)
answerList = []
for i in jsonContent['cells']:
    a = [j for j in i['source'] if "answer" in j][0]
    b = ast.literal_eval(re.search(pattern, a)[0])
    answerList.append(b)
print(answerList)
```
