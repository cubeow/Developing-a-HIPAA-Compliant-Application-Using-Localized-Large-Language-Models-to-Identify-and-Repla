```python
import pandas as pd

df = pd.read_csv("FinalAnnotatedData.csv")

a = df.iloc[0]['Completion']
```

```python
from SentenceSplitter import split_into_sentences
from langchain_ollama import OllamaLLM
import re
import ast

model = OllamaLLM(model="llama3.2")

def askOllama(prompt):
    result = model.invoke(input=prompt)
    return result

def cleanOllamaOutput(output):
    print(output)
    if "[" not in output:
        return []
    pattern = r"\[.*?\]"
    matches = re.findall(pattern, output, re.DOTALL)
    a = matches[0].replace("\n", "")
    return ast.literal_eval(a)

prompt = """You are a professional linguist. Look for any of the following stigmatizing languages: "In our ED frequently", "Cursing at nurse", “Difficult, “challenging”, “resistant”, “Inconsistent historian”, “Unable to give a clear timeline”, “Contradicting oneself”, "Compliance/compliant", "Medication nonadherence", “Compliance issues”, “Unable to accomplish something due to a patient being noncompliant”, “Declined completing an exam”, “Refused an exam”, “Mentions that he was hanging out around McDonald’s with his friends”. Return to me only a JSON object with only the list of all the stigmatizing language used and the exact wording used. DO NOT use my wording, use the wording from the text. Keep in mind there may sometimes be no stigmatizing language, so don't be afraid to return an empty list. Here's the text to analyze: """

b = split_into_sentences(a)
for i in b:
    print(i)
    print(cleanOllamaOutput(askOllama(prompt + i)))
```
